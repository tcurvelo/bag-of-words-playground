```python
import pandas as pd
import re

data = pd.read_csv('./data/jurisprudencia.csv')
data.rename(index=str, columns={'TEMA': 'label', 'EMENTA': 'text'}, inplace=True)
data.sort_values('label').head(15)
```

Os temas possuem especificidades ("subtemas"). Vamos ignorá-las por ora, apagando do hífen em diante.

```python
data['label'] = data['label'].apply(lambda i: re.sub(' +-.*', '', i)).astype(str)
data.head()
```

```python
data.groupby('label').agg('count').sort_values(by='text', ascending=False).head(10)
```

```python
data.groupby('label').agg('count').sort_values(by='text', ascending=False).head()
```

```python
import featx
labeled_feature_set = featx.label_feats_from_dataframe(data)
```

```python
train_feats, test_feats = featx.split_label_feats(labeled_feature_set, 0.8)
nb_classifier = featx.NaiveBayesClassifier.train(train_feats)
print(featx.accuracy(nb_classifier, test_feats))
```
